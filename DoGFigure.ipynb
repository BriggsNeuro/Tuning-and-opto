```python
import numpy as np
import matplotlib.pyplot as plt


# Define grid; arbitrary numbers; create grid size as you'd like
x, y = np.meshgrid(np.linspace(-7, 7, 100), np.linspace(-7, 7, 100))  # Grid size

# Parameters for first DoG plot, adjust to appearance
sigma1_1 = 1.5  # Standard deviation (first Gaussian); 
sigma2_1 = 7  # Standard deviation (second Gaussian)
mask1 = 6  # mask for brim

# Parameters for second DoG plot 
sigma1_2 = 2  
sigma2_2 = 9  
mask2 = 6  

# Compute radial distance / Euclidean distance (2D coordinates)
r = np.sqrt(x**2 + y**2) 

# Compute the Difference of Gaussians, for 3D
# Gaussian distribution equation: https://www.geeksforgeeks.org/python-gaussian-fit/ 
z1 = (np.exp(-r**2 / (2 * sigma1_1**2)) - np.exp(-r**2 / (2 * sigma2_1**2)))
z2 = (np.exp(-r**2 / (2 * sigma1_2**2)) - np.exp(-r**2 / (2 * sigma2_2**2)))

# Apply masks to limit the brim to a circle; because the edges would be square 
z1[r > mask1] = np.nan
z2[r > mask2] = np.nan

# Plot
fig = plt.figure(figsize=(12, 6))

# First subplot 
# 3D Plot: https://matplotlib.org/stable/plot_types/3D/surface3d_simple.html 
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(x, y, z1, cmap='coolwarm')#adjustable, python colormap options: https://matplotlib.org/stable/users/explain/colors/colormaps.html
ax1.grid(True)  # Enable grid
ax1.set_title('Modified')

# Second subplot )
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(x, y, z2, cmap='coolwarm')
ax2.grid(True)  # Enable grid
ax2.set_title('Original')

plt.tight_layout()
plt.show()

```
